```python
# Alt version
from transformers import AutoProcessor, MusicgenForConditionalGeneration
processor = AutoProcessor.from_pretrained("facebook/musicgen-stereo-small")
model = MusicgenForConditionalGeneration.from_pretrained("facebook/musicgen-stereo-small")
```

```python
inputs = processor(
    text=["Drum and Bass"],
    padding=True,
    return_tensors="pt",
)
```

```python
audio_values = model.generate(
    **inputs, 
    max_new_tokens=256,
    temperature=1,
    top_p=0,
    top_k=500,
    guidance_scale=5
)
```

```python
import os
import soundfile as sf
audio_file_path = os.path.join('./', f"test.wav")
sampling_rate = model.config.audio_encoder.sampling_rate
audio_values = audio_values.cpu().numpy()
sf.write(audio_file_path, audio_values[0].T, sampling_rate)
```

```python
from audiocraft.models import MusicGen
from audiocraft.data.audio import audio_write

model = MusicGen.get_pretrained("facebook/musicgen-stereo-small", device='cpu')
model.set_generation_params(
    duration=10,
    temperature=1,
    top_k=250,
    top_p=0,
    cfg_coef=5
)  # generate 8 seconds.

descriptions = ["drum and bass"]

wav = model.generate(descriptions)  # generates 2 samples.

for idx, one_wav in enumerate(wav):
    # Will save under {idx}.wav, with loudness normalization at -14 db LUFS.
    audio_write(f'{idx}', one_wav.cpu(), model.sample_rate, strategy="loudness")
```
